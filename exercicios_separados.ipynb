```python
import requests
from datetime import datetime
```

```python
URL = 'https://api.itjobs.pt/job/list.json?api_key=9fa7ce317d6e85c90d92244adb9146c6'
request = requests.get(URL,headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/118.0'})
data = request.json()
data
```

```python
# Exercício a)

lst = {}
for i in range(0,12):
    b = data['results'][i]['title']
    a = data['results'][i]['publishedAt']
    a = datetime.strptime(a, "%Y-%m-%d %H:%M:%S")
    lst[b] = a
#print(lst)

lstsorted = sorted(lst.items(), key=lambda x: x[1])

for key, value in lstsorted:
    print(f'{key}: {value}')
```

```python
# Exercício b)

lst = []
for i in range(0,12):
    try:
        print(data['results'][i]['types'][0]['name'])
        lst.append(data['results'][i]['id'])
        print(lst)
    except:
        continue
```

```python
jobid = input('Digite o id do trabalho: ')

# Presume-se que 'data' seja um dicionário contendo os resultados.
for i in range(0, 12):
    if data['results'][i]['id'] == jobid:
        a = data['results'][i]['wage']
        
        if a is None:
            if 'contracts' in data['results'][i] and 'types' in data['results'][i]:
                print(f'Sem informação sobre o salário.\nO contrato: {data["results"][i]["contracts"][0]["name"]}, {data["results"][i]["types"][0]["name"]}')
            elif 'contracts' in data['results'][i]:
                print(f'Sem informação sobre o salário.\nO contrato: {data["results"][i]["contracts"][0]["name"]}')
            elif 'types' in data['results'][i]:
                print(f'Sem informação sobre o salário.\nO contrato: {data["results"][i]["types"][0]["name"]}')
            else:
                print('Sem informação sobre o trabalho selecionado.')
        else:
            print(f'Apresenta um salário inicial de {a}€.')
        
        break  # Saia do loop ao encontrar o trabalho com o id correspondente
else:
    print('Trabalho não encontrado.')
```
